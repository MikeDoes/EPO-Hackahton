```python
GDRIVE_PATH='/content/drive/MyDrive/Hackathon: European Patent Office (EPO)/data/'



gpt3_training_file = []

with open(GDRIVE_PATH+'relation_extraction_dataset.json', 'r') as file:
    dataset = json.load(file)      
    for element in dataset:
      gpt3_training_file += [
          {"prompt":element[0][:800],
            "completion":element[1]}
          ]

json.dump(gpt3_training_file, open(GDRIVE_PATH + f"open_ai_training.json", "w"))
```

```python
import json

# Read the JSON file
with open(GDRIVE_PATH + f"open_ai_training.json", "r") as json_file:
    data = json.load(json_file)

# Write the JSONL file
with open(GDRIVE_PATH + f"open_ai_training.jsonl", "w") as jsonl_file:
    for item in data:
        item['prompt'] = "Patent Claims:\n" + item['prompt'][:800] + '\nExtract a small knowledge graph from the patent claims in the format of a complete json list (max 10 triplets) of ["entity_1", "relation", "entity_2"], ... :'
        item['completion'] = json.dumps(item['completion'])
        json.dump(item, jsonl_file)
        jsonl_file.write('\n')
        break
```
